# **Introduction**

Naive Bayes is a simple probabilistic classifier based on applying Bayes' theorem with strong (naive) independence assumptions between the features.

Naive Bayes classifiers are a popular statistical technique of e-mail filtering. They typically use bag of words features to identify spam e-mail, an approach commonly used in text classification.

The Naive Bayes algorithm is based on the Bayes theorem, which states that the probability of an event occurring is equal to the prior probability of the event occurring, multiplied by the likelihood of the event occurring given the evidence.

In a Naive Bayes classifier, each feature is treated as being independent of the other features. This assumption is called class conditional independence. Given this assumption, the probability of a data point belonging to a particular class can be calculated by multiplying the probabilities of each individual feature belonging to that class.

The Naive Bayes algorithm is a fast and simple algorithm, and it works well on many real-world datasets. However, it has a number of limitations, including the assumption of independent features and the need to estimate the probability of each class and each feature value. Despite these limitations, the Naive Bayes algorithm can still be a useful tool in certain situations.

```python
## cleaning the text

def cleantext(text):
    
    # removing the "\"
    
    text = re.sub("'\''","",text)
    
    # removing special symbols
    
    text = re.sub("[^a-zA-Z]"," ",text)
    
    # removing the whitespaces
    
    text = ' '.join(text.split())
    
    # convert text to lowercase
    
    text = text.lower()
    
    return text


    
books['summary'] = books['summary'].apply(lambda x:cleantext(x))
books['summary'].iloc[1]
```

**Plotting the most Frequent Words**

```python
def showmostfrequentwords(text,no_of_words):
    
    allwords = ' '.join([char for char in text])
    allwords = allwords.split()
    fdist = nltk.FreqDist(allwords)
    
    wordsdf = pd.DataFrame({'word':list(fdist.keys()),'count':list(fdist.values())})
    
    df = wordsdf.nlargest(columns="count",n = no_of_words)
    
    plt.figure(figsize=(10,9))
    ax = sn.barplot(data=df,x = 'count',y = 'word', palette='husl')
    ax.set(ylabel = 'Word')
    plt.show()
    
    return wordsdf
    
    
# 30 most frequent words

wordsdf = showmostfrequentwords(books['summary'],30)
```

```python
wordsdf.sort_values('count',ascending=False).head(10).style.background_gradient(cmap = 'plasma')
```

**Removing the stopwords**

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

# removing the stopwords

def removestopwords(text):
    
    removedstopword = [word for word in text.split() if word not in stop_words]
    return ' '.join(removedstopword)

books['summary'] = books['summary'].apply(lambda x:removestopwords(x))
books['summary'].iloc[1]
```

### **Stemming**

```python
# Approach 

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

def stemming(sentence):
    
    stemmed_sentence = ""
    for word in sentence.split():
        stem = stemmer.stem(word)
        stemmed_sentence+=stem
        stemmed_sentence+=" "
        
    stemmed_sentence = stemmed_sentence.strip()
    return stemmed_sentence

books['summary'] = books['summary'].apply(lambda text:stemming(text))
books['summary'].iloc[1]
```

### **Encoding**

```python
books_list = list(books['genre'].unique())
encode = [i for i in range(len(books_list))]
mapper = dict(zip(books_list,encode))
print(mapper)
```

```python
books['genre'] = books['genre'].map(mapper)
books['genre'].unique()
```

# **Model Building**

There are several variations of the Naive Bayes algorithm depending on the distribution of the data. These variations of the Naive Bayes algorithm include;

1. Gaussian: This type of classifier assumes the distribution of features to be Gaussian or normal. This is the most common type of Naive Bayes algorithm that we will come across. Thus, we will perform more challenges using this type of Naive Bayes classifier in today's exercise. 

2. Multinomial: This type of classifier is used when the data is distributed multinomially, i.e., multiple occurrences matter a lot.

3. Bernoulli: This type of classifier is used when the features in the data set are binary-valued


For this case study will be using Multinomial algorithm

```python
## count vectorizer

count_vec = CountVectorizer(max_df=0.90,min_df=2,
                           max_features=1000,stop_words='english')

bagofword_vec = count_vec.fit_transform(books['summary'])
bagofword_vec
```

```python
#spliting train and test 80, 20 
test = books['genre']
X_train, X_test, y_train, y_test = train_test_split(bagofword_vec,test,
                                                    test_size=0.2)
X_train.shape,X_test.shape
```

```python
#Building model
mb = MultinomialNB()
mb.fit(X_train,y_train)
mbpred = mb.predict(X_test)
print(metrics.accuracy_score(y_test,mbpred))
```

## **Part 2 Model Building**

Changing from Countvectorizer to TFDIF vectorizer

```python
#Labeling each 'genre' with an unique number 

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y=le.fit_transform(books['genre'])

X_train,X_test,y_train,y_test = train_test_split(books['summary']
                                                ,y,test_size=0.2,
                                                random_state=557)

X_train.shape,X_test.shape
```

```python
#Performing tf-idf 

tfidf_vectorizer = TfidfVectorizer(max_df=0.8, max_features=10000)
xtrain_tfidf = tfidf_vectorizer.fit_transform(X_train.values.astype('U'))
xtest_tfidf = tfidf_vectorizer.transform(X_test.values.astype('U'))
```

```python
mb = MultinomialNB()
mb.fit(xtrain_tfidf,y_train)
mbpred = mb.predict(xtest_tfidf)
print(metrics.accuracy_score(y_test,mbpred))
```

### **Testing the Model**

```python
def test(text,model):
    
    text = cleantext(text)
    text = removestopwords(text)
    #text = lematizing(text)
    text = stemming(text)
    
    text_vector = tfidf_vectorizer.transform([text])
    predicted = model.predict(text_vector)
    return predicted



ans = books['summary'].apply(lambda text:test(text,mb))

# printing the 
# print(list(mapper.keys())[list(mapper.values()).index(ans)])

ans
```

```python
predicted_genres = []
for i in range(len(ans)):
    
    index_val = ans[i][0]
    predicted_genres.append(list(mapper.keys())[list(mapper.values()).index(index_val)])
    
```
